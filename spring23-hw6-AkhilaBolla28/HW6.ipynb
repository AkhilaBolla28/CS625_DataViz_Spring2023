```python
# Dropping the unwanted columns
df_1.drop("App Id",axis=1, inplace=True)
df_1.drop("Minimum Installs",axis=1, inplace=True)
df_1.drop("Maximum Installs",axis=1, inplace=True)
#df_1.drop("Free",axis=1, inplace=True)
df_1.drop("Currency",axis=1, inplace=True)
df_1.drop("Minimum Android",axis=1, inplace=True)
df_1.drop("Developer Id",axis=1, inplace=True)#
df_1.drop("Developer Website",axis=1, inplace=True)
df_1.drop("Developer Email",axis=1, inplace=True)
df_1.drop("Released",axis=1, inplace=True)
df_1.drop("Last Updated",axis=1, inplace=True)
#df_1.drop("Content Rating",axis=1, inplace=True)
df_1.drop("Privacy Policy",axis=1, inplace=True)
df_1.drop("Ad Supported",axis=1, inplace=True)
df_1.drop("In App Purchases",axis=1, inplace=True)
df_1.drop("Editors Choice",axis=1, inplace=True)
df_1.drop("Scraped Time",axis=1, inplace=True)
df_1.drop("Rating Count",axis=1, inplace=True)
```

```python
# Displaying to get the first 4 rows

df_1.head()
```

```python
#Checking for Null values in the data set
df_1.isnull().sum()
```

```python
#Rating column
df_1['Rating'].fillna(df_1['Rating'].mean(),inplace=True)

#App Name
df_1['App Name'].fillna(method='ffill',inplace=True)

#Size
df_1['Size'].fillna(method='ffill',inplace=True)

#Installs
df_1['Installs'].fillna(method='ffill',inplace=True)
```

```python
#Checking for Null values in the data set
df_1.isnull().sum()
```

```python
# Verifying the Maximum count in Rating attribute

df_1['Rating'].max()
```

```python
# Removing the special characters from the Install column 
df_1['Install1'] = df_1["Installs"].str.replace('+','')
df_1['Install']=df_1['Install1'].str.replace(',','')
df_1.Size = [x.strip().replace('M', '') for x in df_1.Size]
df_1.Size = [x.strip().replace('k', '') for x in df_1.Size]
```

```python
## Changing few categorical to numerical

df_1["Rating"]= df_1.Rating.apply(lambda x: int(x))
df_1["Price"] = df_1.Price.apply(lambda x: float(x))
df_1['Install']= [int(x) if x.isnumeric() else x for x in df_1['Install']]
```

### **Visualization - Android Data**

```python
#Number of categories of apps in the store.
def plot_number_category():
    fig, ax = plt.subplots()
    fig.set_size_inches(15, 7)
    fig.autofmt_xdate()
    countplot=sns.categorical.countplot(df_1.Category,ax=ax)
    plt.xticks(rotation = 90)
    plt.show(countplot)

plot_number_category()

# Tabular representation
top_cat=df_1.groupby('Category').size().reset_index(name='Count').nlargest(5,'Count')
display(top_cat)
```

The above data shows that the top most category is Education followed by Music&

Audio , Tools, Bussiness and Entertainment.


```python
top5_cat = top_cat['Category'].tolist()
data_top5 = df_1.groupby('Category')['Install'].agg('sum').loc[top5_cat].reset_index(name='Number_Installations')
data =  df_1.groupby('Category')['Install'].agg('sum').reset_index(name='Number_Installations')
```

```python
#Comparing top 5 category on the basis of Installs
fig = plt.figure(figsize=(10,5))
title=plt.title('Comparing top 5 category on the basis of Installs')
bar=sns.barplot(y=data_top5['Category'],x=data_top5['Number_Installations'])
plt.show(bar)
```

```python
##Comparing top 5 category on the basis of Content Rating

data_cont =  df_1.groupby('Content Rating')['Install'].agg('sum').reset_index(name='Number_Installations')
plt.figure(figsize=(10,5))
sns.barplot(x=data_cont['Content Rating'], y=data_cont['Number_Installations'])
plt.show()
```
